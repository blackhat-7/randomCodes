```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.arange(1, 10, 0.01)
y1 = np.log2(x)
y2 = np.log2(np.log2(x))
y3 = x
y4 = x * np.log2(x)
y5 = x * np.log2(np.log2(x))
y6 = x**2
y7 = x * np.sqrt(x)

plt.figure(figsize=(7,3), dpi=150)

plt.plot(x, y1, label='LogN', color="yellow")
plt.plot(x, y2, label='LogLogN', color="black")
plt.plot(x, y3, label='N', color="blue")
plt.plot(x, y4, label='NLogN', color="green")
plt.plot(x, y5, label='NLogLogN', color="red")
# plt.plot(x, y6, label='N^2', color="purple")
plt.plot(x, y7, label='Nsqrt(N)', color="pink")


plt.title('First Plot', fontdict={'fontname': 'Monaco', 'fontsize': 20})

plt.xlabel('X Axis', fontdict={'fontname': 'Monaco', 'fontsize': 13})
plt.ylabel('Y Axis', fontdict={'fontname': 'Monaco', 'fontsize': 13})

# plt.xticks(np.arange(0, 5, 0.5))
# plt.yticks(np.arange(0, 0.5, 0.5))

plt.legend()
plt.savefig('TimeComplexities.png', dpi=300)
plt.show()
```
