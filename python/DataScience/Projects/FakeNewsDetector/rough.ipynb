```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./data/train.csv")
df.head()
```

```python
df["total"] = df["title"] + " " + df["author"] + " " + df["text"]
```

```python
stopwords = set(STOPWORDS)
real_words = ""
fake_words = ""

for row in df[["total", "label"]].values:
    row[0] = str(row[0])
    if row[1] == 1:
        real_words += row[0] + " "
    else:
        fake_words += row[0] + " "
wordcloud_real = WordCloud(stopwords=stopwords).generate(real_words)
wordcloud_fake = WordCloud(stopwords=stopwords).generate(fake_words)
```

```python
plt.imshow(wordcloud_real)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
plt.imshow(wordcloud_fake)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
