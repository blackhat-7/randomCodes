```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='ISO-8859-1')
df.head()
```

```python
sns.lmplot(data=df, x='Attack', y='Defense', fit_reg=False, hue='Stage')
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sns.boxplot(data=df)
```

```python
status_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=status_df)
```

```python
sns.violinplot(data=status_df, x='Type 1', y='Attack')
```
