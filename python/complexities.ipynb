```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
import plotly.express as px
plt.style.use('ggplot')


N = np.arange(1.1, 10, 0.01)

fns = {
    'N': lambda N: N,
    'log(N)': lambda N: math.log(N),
    'log(log(N))': lambda N: math.log(math.log(N)),
    'sqrt(N)': lambda N: math.sqrt(N),
    'Nlog(N)': lambda N: N*math.log(N),
    'Nlog(log(N))': lambda N: N*math.log(math.log(N)),
    'Nsqrt(N)': lambda N: N*math.sqrt(N),
    # 'N^2': lambda N: N**2,
    # '2^N': lambda N: 2**N,
    # 'N!': lambda N: math.factorial(N),
}
```

```python
complexities = pd.DataFrame()
complexities['points'] = N
for fn in fns:
    complexities[fn] = [fns[fn](n) for n in N]
complexities
```

```python
fig = px.line(
    data_frame=complexities,
    x='points',
    y=complexities.columns
)
fig.update_layout(title='Order Complexities Comparison')
fig.show()
```
