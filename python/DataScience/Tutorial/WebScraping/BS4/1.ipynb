```python
def get_prices(soup, prices):
    print(len(prices))
    for product in soup.find_all('div', class_='sg-col-inner'):
        title = product.find('span', attrs={'class':'a-size-medium a-color-base a-text-normal', 'dir': 'auto'})
        if title:
            title = title.get_text()
        else:
            print(None)
            continue
#         price = product.find('span', class_='a-price-whole')
#         if price:
#             price = float(price.get_text().replace(',','') + product.find('span', class_='a-price-fraction').get_text())
#         else:
#             continue
        prices[title] = 0
```

```python
browser = webdriver.Safari()
browser.implicitly_wait(10)
browser.get("https://www.amazon.com/s?k=macbook&rh=n%3A172282%2Cp_89%3AApple&dc&qid=1607644508&rnid=2528832011&ref=sr_nr_p_89_1")
soup = bs(browser.page_source, 'lxml')
ui.WebDriverWait(browser, 10).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, ".a-size-medium.a-color-base.a-text-normal")))
prices = {}
get_prices(soup, prices)
for i in range(9): 
    browser.find_element_by_css_selector('.a-last').click()
    soup = bs(browser.page_source, 'lxml')
    ui.WebDriverWait(browser, 10).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, ".a-size-medium.a-color-base.a-text-normal")))
    get_prices(soup, prices)
browser.quit()
len(prices)
```
