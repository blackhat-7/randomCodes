# Line Objects

```python
import plotly.graph_objects as go
df_stocks = px.data.stocks()
px.line(df_stocks, x='date', y='GOOG', labels={'x':'Date', 'y':'Price'})
px.line(df_stocks, x='date', y=['GOOG', 'AAPL'], labels={'x':'Date', 'y':'Price'}, title='Google Vs Apple Stocks')
fig = go.Figure()
fig.add_trace()
```
